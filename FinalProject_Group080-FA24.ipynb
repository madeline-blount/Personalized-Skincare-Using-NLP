**Data Wrangling: Merging the Datasets, and Turning a String Variable into a List Variable**

In the following code we cleaned the "highlights" variable in the product information column. This column includes basic information or "highlights" about what the product is supposed to be used for. In the original dataset, this is given as a string, but we decided to convert this to a list, with each feature of the item it mentions as a separate item in the list, which will make it easier for us count highlights, and to look at how many products share a certain highlight. To do this we removed special characters and separated items by commas, then appended each new phrase or word to an empty list. In our updated data frame, there is a list for each observation under that variable. 

In this code block, we also merged the two datasets based on the shared product ID variable, so that our new dataframe, merged_df, includes one row for each user review, as well as important product information, such as ingredient list and proudct highlights. 

```python
import numpy as np

merged_df = pd.merge(data, all_reviews, how='inner', on='product_id')
merged_df.sort_values('average_rating')
merged_df['highlights'].unique()

def clean_highlights(highlights):
    if type(highlights) == float: return np.nan
    string = highlights.strip("[]")
    elements = string.split(", ")
    elements = [element.strip("'") for element in elements]
    good_fors = []
    for element in elements:
        if 'Good for:' in element:
            element = element.split("Good for: ")[1]
            good_fors.append(element)
    if good_fors: return good_fors
    return np.nan

merged_df['highlights'] = merged_df['highlights'].apply(clean_highlights)
merged_df = merged_df[~merged_df['highlights'].isna()]
```

```python
merged_df = merged_df.rename(columns={'highlights': 'good_for'})
merged_df
```

**Data Cleaning:** To ensure that the ingredients list of each product did not have inconcistencies, we removed characters like * and made sure that each ingredient is separated by a comma.

```python
import re

def clean_ingredients(ingredients):
    if type(ingredients) == float: return np.nan
    ingredients = re.split(r',\s*(?![^()]*\))', ingredients.strip("[]"))
    return [ingredient.replace("'", '').replace('*', '').rstrip('.').strip() for ingredient in ingredients]

clean_df = merged_df.copy()

clean_df['ingredients'] = clean_df['ingredients'].apply(clean_ingredients)

clean_df = clean_df.dropna(subset=['ingredients']).reset_index(drop=True)
```

```python
def convert_to_list(value):

    if isinstance(value, str):
        value = value.strip('[]')
        return [elem.strip().strip("'") for elem in value.split(',')]
    elif isinstance(value, list):
        return value
    else:
        return []
```

The primary input feature for our model, good_for_tfidf, is a vectorized representation of the good_for column, which captures the skin concerns associated with each product. The output, good_for_binarized, is a binary matrix where rows represent products and columns correspond to specific skin concerns, with a value of 1 indicating the presence of a concern and 0 indicating its absence. The model predicts the suitability of a product for each skin concern based on its associated features. This multi-label classification framework allows us to handle products labeled with multiple skin concerns, enabling predictions across all concerns simultaneously.

```python
# training model

def train_model(product_features_df):
    product_features_df = product_features_df.copy()

    # convert good_for column to a list of lists
    product_features_df['good_for'] = product_features_df['good_for'].apply(convert_to_list)
    product_features_df['good_for'] = product_features_df['good_for'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)

    # vectorize good_for
    tfidf = TfidfVectorizer(max_features=1000, tokenizer=lambda x: x.split(', '))
    good_for_tfidf = tfidf.fit_transform(product_features_df['good_for'])

    # binarize labels for multi-label classification
    mlb = MultiLabelBinarizer()
    products_binarized = mlb.fit_transform(product_features_df['good_for'].apply(lambda x: [x]))

    print("Number of unique products:", len(mlb.classes_))
    print("Products:", mlb.classes_)

    # train-test split
    X_train, X_test, y_train, y_test = train_test_split( 
        good_for_tfidf, 
        products_binarized, 
        test_size=0.2, 
        random_state=42
    )
    
    # train model using separate binary logistic regression models for each concern
    model = OneVsRestClassifier(LogisticRegression(max_iter=1000))
    model.fit(X_train, y_train)

    return model, tfidf, mlb, merged_df, X_test, y_test, X_train, y_train


model, tfidf, mlb, merged_df, X_test, y_test, X_train, y_train = train_model(transformed_df)
```

The function, recommend_products_for_concern allows us to apply the model defined above to the specific skin concerns of the user, by using MultiLableBinarizer (mlb) to map the user’s concerns to the classification of the product and finding the probability that a product will suit a user’s concern. In function creates a matrix where each row represents a product, and each column a concern, then using this matrix is able to see how well the product satisfies the input concerns of the user. It outputs the top five products and includes relevant information, like average user reviews and average sentiment score.

```python
def recommend_products_for_concern(model, X_test, mlb, concern, threshold=0.5, merged_df=None, top_n=5):
    # Ensure merged_df is provided
    if merged_df is None:
        raise ValueError("merged_df is required to fetch product names.")
    
    # Ensure the concern is a valid skin concern in the model
    if concern not in mlb.classes_:
        raise ValueError(f"{concern} is not a valid skin concern. Choose from {mlb.classes_}")
    
    # Get the predicted probabilities for each skin concern
    y_pred_proba = model.predict_proba(X_test)
    
    # Find the indices of products that are predicted to be suitable for the specified concern
    concern_idx = np.where(mlb.classes_ == concern)[0][0]  # Index of the specific concern
    recommended_indices = np.where(y_pred_proba[:, concern_idx] >= threshold)[0]
    
    # Get the product names and their average ratings corresponding to the indices
    recommended_products_df = merged_df.iloc[recommended_indices][['product_name', 'average_rating', 'sentiment_score']]
    
    # Sort the products by the average rating in descending order
    recommended_products_sorted = recommended_products_df.sort_values(by='average_rating', ascending=False)
    top_recommended_products = recommended_products_sorted.head(top_n)

    # top_recommended_products = top_recommended_products.reset_index(drop=True)
    # top_recommended_products.index = top_recommended_products.index + 1


    return top_recommended_products
```

In order to see how well our model recommended relevant products, we compared the product classifications created in our model with the actual product labels. 

The precision score helps us see how specific our model is in predicting a truly relevant product. The more specific, the more likely the model is to predict false negatives. To evaluate precision, the model divides the probability of a true positive result over the combined probability of true and false positives. Using this formula, a lower precision score indicates less precision, and the model is more likely to result in false positives. 

Conversely, the recall score helps us identify how sensitive our model is, the more sensitive a model is, the more likely it is to result in false positives. For example, a lower recall score would indicate that a model is less sensitive, and is less likely to positively classify a product, resulting in false negatives.  

The F1-Score is a harmonic mean between the precision score and the recall score, meaning it balances how small or large either of those values is. It is another measure of how accurate the model is, combining precision and recall to get a more balanced measure of the accuracy. 

The support value represents the number of products in the test set that fit in the category of the respective row. The larger the support number is, the more instances there are to back up the accuracy statistics, and the more confident we can feel about the model’s accuracy measurements.

```python
from sklearn.metrics import classification_report, accuracy_score

y_pred = model.predict(X_test)
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=mlb.classes_))
```

In our final results, sentiment analysis revealed key trends in customer satisfaction, with higher sentiment scores often linked to positive experiences and product effectiveness for specific concerns. Among the skin concerns analyzed, “Dullness/Uneven Texture” emerged as one of the most common, demonstrated by a high support value. We then identified the top five recommended products for “Dullness/Uneven Texture” using our model and sorting by average ratings. The top-ranked product was the ”Pumpkin Enzyme Mask Enzymatic Dermal Resurfacer”, with a sentiment score of 0.74, followed by the Confidence in a “Cream Anti-Aging Hydrating Moisturizer”, which had the highest sentiment score of 0.81. Both products stood out for their strong alignment with user preferences and high ratings. Other top recommendations included ”Alpha Beta Extra Strength Daily Peel Pads” (Average Rating: 4.5455, Sentiment Score: 0.651869), ”FIRMx Exfoliating Peel Gel” (Average Rating: 4.5391, Sentiment Score: 0.524774), and ”Therapeutic Sulfur Acne Treatment Mask” (Average Rating: 4.5111, Sentiment Score: 0.591516), all of which demonstrated strong potential for addressing “Pores” effectively.

The model struggled with skin concerns that had less data such as “Damage,” which affected its accuracy in those areas. The model also has limited capabilities in terms of the features that it can take into its inputs. We struggled with incorporating more complex data such as the ingredients of the products into our model due to the inconsistencies within the actual data. Future improvements include combining various features of our dataset to strengthen the accuracy of our model. 

While this project shows the value of using data science to recommend skincare products, it also raises ethical issues. The reviews we used were public and anonymized, but users might not fully understand how their data is being used, even if they agreed to Sephora’s terms. Furthermore, as demonstrated by our heavily left-skewed frequency graph for user review sentiment scores, reviews tended to show extreme opinions, either very positive or very negative, with few neutral opinions. Because our data is made of people who chose to leave a review, rather than a random selection of people who have bought a certain product, our dataset suffers from self-selection bias.
